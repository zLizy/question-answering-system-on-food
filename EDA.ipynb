# Clean the data and make some of the columns from string to float

```python
df_clean = pd.DataFrame()
columns = df.columns
for col in columns:
    try:
        df_clean[col] = df[col].str.replace(',', '').astype(float)
        df_clean[col].replace(np.nan, 0, inplace=True)
        print(f'change {col} from {df[col].dtypes} to {df_clean[col].dtypes}')
    except (ValueError, AttributeError) as e:
        df_clean[col] = df[col]
```

```python
df_clean.dtypes
```

```python
df_clean['unit'] = 100
```

```python
df_clean.to_csv('NEVO2023_8.0_clean.csv')
```

```python
# Filter for the "Groente" category and calculate the average fat content
avg_fat_content_vegetables = df_clean[df_clean['Voedingsmiddelgroep'] == "Groente"]['FAT (g)'].mean()

# Print the result
print(avg_fat_content_vegetables)
```

```python
import pandas as pd
# Load the dataset
df = pd.read_csv('NEVO2023_8.0_clean.csv')
# Filter for white asparagus
white_asparagus = df[df['Voedingsmiddelnaam/Dutch food name'] == 'Asperge witte rauw']
# Extract vitamin C content per 100 grams
vitamin_c_content = white_asparagus['VITC (mg)'].iloc[0] # Assuming this column contains total vitamin C content
print(vitamin_c_content)
# Calculate vitamin C content in 200 grams
vitamin_c_200g = vitamin_c_content * 2
# Print result
print(f'The consumption of 200 grams of white asparagus fulfills {vitamin_c_200g / 90 * 100:.2f}% of the daily vitamin C requirement.')
```
